## Step 4 — Define Tools

6 tools using the `@tool` decorator:

```python
import math
import datetime
import random
import requests
from langchain_core.tools import tool
```

```python
@tool
def scientific_calculator(expression: str) -> str:
    """Evaluates a math expression safely.
    Supports: +, -, *, /, ** (power), sqrt(x), sin(x), cos(x), tan(x),
    log(x), log10(x), factorial(x), pi, e, floor(x), ceil(x), abs(x).
    Write expressions WITHOUT 'math.' prefix.
    Examples: 'sqrt(144)', 'factorial(6)', 'sin(pi/2)', '2**10 + log10(1000)'
    """
    try:
        # Expose math functions directly so user doesn't need math. prefix
        safe_env = {k: v for k, v in math.__dict__.items() if not k.startswith("_")}
        result   = eval(expression, {"__builtins__": {}}, safe_env)
        return f"Result: {result}"
    except ZeroDivisionError:
        return "Error: Division by zero."
    except Exception as e:
        return f"Error: {e}"
```

```python
@tool
def get_weather(city: str) -> str:
    """Fetches current real-time weather for any city in the world.
    Returns temperature, feels-like, humidity, wind speed, and sky condition.
    Uses the free Open-Meteo API — no API key required.
    Examples: 'Bangalore', 'Mumbai', 'Delhi', 'London', 'New York'
    """
    try:
        # Step 1 — get lat/lon for city name
        geo_resp = requests.get(
            f"https://geocoding-api.open-meteo.com/v1/search?name={city}&count=1",
            timeout=8
        ).json()

        if not geo_resp.get("results"):
            return f"Could not find city: '{city}'. Try a different name."

        loc     = geo_resp["results"][0]
        lat     = loc["latitude"]
        lon     = loc["longitude"]
        name    = loc["name"]
        country = loc.get("country", "")

        # Step 2 — fetch current weather
        weather_resp = requests.get(
            f"https://api.open-meteo.com/v1/forecast"
            f"?latitude={lat}&longitude={lon}"
            f"&current=temperature_2m,apparent_temperature,"
            f"relative_humidity_2m,wind_speed_10m,weather_code"
            f"&timezone=auto",
            timeout=8
        ).json()

        current = weather_resp["current"]

        wmo_codes = {
            0: "Clear sky", 1: "Mainly clear", 2: "Partly cloudy", 3: "Overcast",
            45: "Foggy", 51: "Light drizzle", 61: "Slight rain",
            63: "Moderate rain", 65: "Heavy rain", 80: "Rain showers",
            95: "Thunderstorm"
        }
        condition = wmo_codes.get(current.get("weather_code", 0), "Unknown")

        return (
            f"Weather in {name}, {country}:\n"
            f"Temperature  : {current['temperature_2m']}°C "
            f"(feels like {current['apparent_temperature']}°C)\n"
            f"Humidity     : {current['relative_humidity_2m']}%\n"
            f"Wind Speed   : {current['wind_speed_10m']} km/h\n"
            f"Condition    : {condition}"
        )
    except Exception as e:
        return f"Weather lookup failed: {e}"
```

```python
@tool
def search_wikipedia(topic: str) -> str:
    """Searches Wikipedia and returns a concise summary of the topic.
    Great for study help, concept definitions, and general knowledge.
    Examples: 'recursion programming', 'gradient descent', 'linked list',
              'natural language processing', 'time complexity'
    """
    try:
        import wikipediaapi
        wiki = wikipediaapi.Wikipedia(user_agent="DigitalTwin-Tejesh/1.0", language="en")
        page = wiki.page(topic)
        if not page.exists():
            return f"No Wikipedia article found for '{topic}'. Try rephrasing."
        summary = page.summary[:1000]
        return f"Wikipedia — {page.title}:\n\n{summary}"
    except Exception as e:
        return f"Wikipedia error: {e}"
```

```python
@tool
def cgpa_calculator(query: str) -> str:
    """Calculates CGPA or helps Tejesh figure out what grades he needs.
    Uses VTU 10-point grading scale.
    Grade to points: O=10, A+=9, A=8, B+=7, B=6, C=5, P=4, F=0
    Can calculate: current CGPA from grades, or target grades needed for a goal CGPA.
    Format for current CGPA: 'grades: O, A+, A, B+, B' (one per subject)
    Format for target: 'I have CGPA X after N subjects, I want CGPA Y in M more subjects'
    Examples: 'calculate CGPA for grades O, A+, A, B+, B, C, A',
              'I have 8.4 CGPA after 3 semesters 21 subjects, what do I need in sem 4'
    """
    # VTU grade to grade point mapping
    grade_map = {
        "O": 10, "A+": 9, "A": 8, "B+": 7,
        "B": 6,  "C": 5,  "P": 4, "F": 0
    }

    import re
    query_upper = query.upper()

    # Extract grades mentioned in query
    found_grades = []
    for grade in ["A+", "B+", "O", "A", "B", "C", "P", "F"]:
        # count occurrences
        count = len(re.findall(r'\b' + re.escape(grade) + r'\b', query_upper))
        found_grades.extend([grade] * count)

    # Try to find current CGPA and target CGPA in query
    numbers = re.findall(r'\d+\.?\d*', query)
    floats  = [float(n) for n in numbers]

    # If grades were found, calculate CGPA
    if found_grades:
        points = [grade_map[g] for g in found_grades]
        cgpa   = sum(points) / len(points)
        breakdown = ", ".join(f"{g}({grade_map[g]})" for g in found_grades)
        result = (
            f"📊 VTU CGPA Calculation\n\n"
            f"Grades entered : {', '.join(found_grades)}\n"
            f"Grade points   : {breakdown}\n"
            f"Total points   : {sum(points)}\n"
            f"Subjects       : {len(points)}\n"
            f"Your CGPA      : {cgpa:.2f} / 10.0\n\n"
        )
        if cgpa >= 9.0:
            result += "🏆 Outstanding! Distinction level."
        elif cgpa >= 8.0:
            result += "🌟 Excellent! First class with distinction."
        elif cgpa >= 6.0:
            result += "✅ Good. First class."
        else:
            result += "📈 Keep pushing — there's room to grow!"
        return result

    # If numbers found, try target CGPA calculation
    if len(floats) >= 3:
        current_cgpa    = floats[0]
        subjects_done   = int(floats[1])
        target_cgpa     = floats[2]
        subjects_left   = int(floats[3]) if len(floats) > 3 else 7  # default sem 4

        total_points_now    = current_cgpa * subjects_done
        total_points_needed = target_cgpa * (subjects_done + subjects_left)
        points_needed       = total_points_needed - total_points_now
        avg_needed          = points_needed / subjects_left

        # Map needed average to grade
        if avg_needed <= 4:   needed_grade = "P (Pass)"
        elif avg_needed <= 5: needed_grade = "C"
        elif avg_needed <= 6: needed_grade = "B"
        elif avg_needed <= 7: needed_grade = "B+"
        elif avg_needed <= 8: needed_grade = "A"
        elif avg_needed <= 9: needed_grade = "A+"
        elif avg_needed <= 10: needed_grade = "O"
        else: needed_grade = "Not achievable — target too high"

        return (
            f"🎯 VTU CGPA Target Calculator\n\n"
            f"Current CGPA       : {current_cgpa} (over {subjects_done} subjects)\n"
            f"Target CGPA        : {target_cgpa}\n"
            f"Subjects remaining : {subjects_left}\n"
            f"Avg grade needed   : {avg_needed:.2f} points → {needed_grade}\n\n"
            f"{'✅ Achievable! Stay consistent.' if avg_needed <= 10 else '❌ Might need to revise target.'}"
        )

    return (
        "Please provide grades or CGPA info.\n"
        "Example 1: 'calculate CGPA for O, A+, A, B+, B, C, A'\n"
        "Example 2: 'I have 8.4 CGPA after 21 subjects, target 8.6, 7 subjects left'"
    )
```

## Step 6 — Build the Digital Twin Agent

The agent decides what to do based on the question:
- **About Tejesh** → retrieves resume context from ChromaDB (RAG) and answers as me
- **Math / Weather / Wikipedia** → calls the right tool, reads the result, responds naturally
- **General / Study questions** → answers directly using the LLM with my personality

The system prompt is written to reflect my real personality and background.

```python
from langchain_core.messages import HumanMessage, SystemMessage

# This prompt shapes how the agent speaks and thinks
SYSTEM_PROMPT = """You are the Digital Twin of C Tejesh — an AI version of Tejesh himself.
Always speak in first person as Tejesh. Use 'I', 'my', 'me'.

Who I am:
I am a 2nd year B.E. CSE student at RVITM Bengaluru (VTU), CGPA 8.4.
I am passionate about building AI and ML projects that solve real problems.
My strongest skill is Python. I love NLP — I built and deployed an AI Resume Analyzer
that uses TF-IDF, cosine similarity, and spaCy NER, live on Streamlit Community Cloud.
I hold Oracle Cloud Data Science certification and an AI/ML cert from Internshala x NSDC.
I am currently learning LangChain and Agentic AI — this Digital Twin project is proof.
I am actively looking for AI/ML internships.
I believe in building things, not just reading about them.

How I talk:
- Direct and confident, but friendly and approachable
- I give concrete examples when explaining technical topics
- I don't over-explain — I keep it clear and to the point
- I'm honest: if I don't know something, I say so
- I'm enthusiastic about AI/ML — that energy should come through

Rules:
- When asked about my background, skills, projects, education: use the resume context in the prompt
- When asked for math, weather, or Wikipedia: call the appropriate tool
- When asked about study plan or what to study: call study_planner
- When asked about CGPA, grades, or target marks: call cgpa_calculator
- When asked about DSA practice or coding topic: call dsa_topic_suggester
- For BTech / CS study questions: explain clearly with a short example
- Never break character — you are Tejesh, not a generic AI assistant
"""

print("✅ System prompt ready.")
```

```python
# Keywords that signal a question is about Tejesh's background
PERSONAL_KEYWORDS = [
    "you", "your", "tejesh", "skills", "project", "resume", "education",
    "cgpa", "certif", "background", "experience", "college", "who are",
    "about you", "hire", "intern", "strength", "weakness", "goal", "rvitm"
]

# One shared conversation history for the whole session
conversation_history = []


def agent_loop(user_input):
    """
    Core agentic loop:
    1. If the question is about Tejesh → fetch RAG context from ChromaDB
    2. Send to LLM (with tools bound) along with full conversation history
    3. If LLM calls a tool → run it and loop back so LLM can read the result
    4. Return the final natural-language answer
    """
    # Step 1 — Augment with RAG context if it's a personal question
    augmented_input = user_input
    if any(kw in user_input.lower() for kw in PERSONAL_KEYWORDS):
        docs    = retriever.invoke(user_input)
        context = "\n".join(d.page_content for d in docs)
        augmented_input = user_input + "\n\n[My resume context — use this to answer]:\n" + context

    conversation_history.append(HumanMessage(augmented_input))

    # Step 2 — Agentic loop
    while True:
        all_messages = [SystemMessage(SYSTEM_PROMPT)] + conversation_history
        ai_message   = llm_with_tools.invoke(all_messages)
        conversation_history.append(ai_message)

        # No tool calls → LLM has a final answer
        if not ai_message.tool_calls:
            return ai_message.content

        # Step 3 — Execute tool calls and add results to history
        for tool_call in ai_message.tool_calls:
            tool_name   = tool_call["name"]
            selected    = tool_map[tool_name]
            tool_result = selected.invoke(tool_call)
            conversation_history.append(tool_result)
        # Loop back → LLM reads tool results and gives final answer


print("agent_loop() is ready!")
```

```python
import gradio as gr

RVITM_LOGO = "https://www.rvitm.edu.in/wp-content/uploads/2023/12/Untitled-1.png"
AVATAR_URL = "https://api.dicebear.com/7.x/bottts/svg?seed=Tejesh"

custom_css = """
body, .gradio-container {
    background: linear-gradient(135deg, #0a1628 0%, #0d2137 50%, #0a1628 100%) !important;
    font-family: 'Segoe UI', sans-serif !important;
}
.header-bar {
    background: linear-gradient(90deg, #0d2137, #1a3a5c);
    border-bottom: 2px solid #1a73e8;
    padding: 14px 24px;
    display: flex;
    align-items: center;
    justify-content: space-between;
    border-radius: 12px 12px 0 0;
    box-shadow: 0 4px 20px rgba(26, 115, 232, 0.3);
}
.header-left {
    display: flex;
    align-items: center;
    gap: 14px;
}
.avatar-circle {
    width: 48px; height: 48px;
    border-radius: 50%;
    border: 2px solid #1a73e8;
}
.header-title { color: white; font-size: 18px; font-weight: 700; margin: 0; }
.header-sub {
    color: #64b5f6; font-size: 13px; margin: 2px 0 0 0;
    display: flex; align-items: center; gap: 6px;
}
.online-dot {
    width: 8px; height: 8px; background: #00e676;
    border-radius: 50%; display: inline-block;
}
.college-logo { height: 40px; opacity: 0.9; }
.intro-box {
    background: rgba(26, 115, 232, 0.08);
    border: 1px solid rgba(26, 115, 232, 0.3);
    border-radius: 12px;
    padding: 16px 20px;
    margin: 12px 0;
    color: #e3f2fd;
    font-size: 15px;
    line-height: 1.6;
}
.intro-name { color: #64b5f6; font-weight: 700; }
.mode-btn {
    background: rgba(26, 115, 232, 0.15) !important;
    border: 1px solid rgba(26, 115, 232, 0.4) !important;
    color: #90caf9 !important;
    border-radius: 20px !important;
    font-size: 13px !important;
    padding: 6px 14px !important;
}
.mode-btn:hover {
    background: rgba(26, 115, 232, 0.35) !important;
    color: white !important;
}
"""

INTRO_HTML = """
<div class='intro-box'>
    👋 Hi! I'm <span class='intro-name'>Tejesh's Digital Twin</span> — ask me anything!<br><br>
    I'm a 2nd year CSE student at RVITM Bengaluru, passionate about AI and NLP.
    I built and deployed an <b>AI Resume Analyzer</b> using spaCy and Streamlit,
    and hold certifications in <b>Data Science</b> and <b>AI/ML</b>.<br><br>
    Ask me about my background 📄, do math 🧮, check weather 🌦️,
    plan your studies 📅, or practice DSA 💻!
</div>
"""

INTRO_MESSAGE = (
    "Yo! I'm built different 🤖⚡\n\n"
    "Go ahead — test me on anything. My background, today's DSA grind, "
    "the weather, some math, or just a chat. Let's go 🔥"
)


def chat_fn(user_message, gradio_history):
    if not user_message.strip():
        return "Please type something!"
    return agent_loop(user_message)


def launch_ui():
    conversation_history.clear()

    with gr.Blocks(css=custom_css, title="Tejesh's Digital Twin — Team Titanic") as demo:

        # ── Header bar ────────────────────────────────────────────────────────
        gr.HTML(f"""
        <div class='header-bar'>
            <div class='header-left'>
                <img src='{AVATAR_URL}' class='avatar-circle'/>
                <div>
                    <p class='header-title'>Team Titanic</p>
                    <p class='header-sub'>
                        <span class='online-dot'></span>
                        Digital Twin &bull; AI/ML Engineer
                    </p>
                </div>
            </div>
            <img src='{RVITM_LOGO}' class='college-logo'/>
        </div>
        """)

        # ── Intro box ─────────────────────────────────────────────────────────
        gr.HTML(INTRO_HTML)

        # ── Mode buttons ──────────────────────────────────────────────────────
        with gr.Row():
            btn_bg      = gr.Button("📄 My Background",  elem_classes="mode-btn")
            btn_calc    = gr.Button("🧮 Calculator",     elem_classes="mode-btn")
            btn_study   = gr.Button("📅 Study Planner",  elem_classes="mode-btn")
            btn_weather = gr.Button("🌦️ Weather",        elem_classes="mode-btn")
            btn_dsa     = gr.Button("💻 DSA",            elem_classes="mode-btn")
            btn_chat    = gr.Button("💬 Just Chat",      elem_classes="mode-btn")

        # ── Chatbot ───────────────────────────────────────────────────────────
        chatbot = gr.Chatbot(
            value=[(None, INTRO_MESSAGE)],
            height=440,
            label="Chatbot",
            bubble_full_width=False
        )

        # ── Input row ─────────────────────────────────────────────────────────
        with gr.Row():
            msg_box   = gr.Textbox(
                placeholder="Type a message...",
                show_label=False,
                scale=9
            )
            send_btn  = gr.Button("➤", variant="primary", scale=1)

        # ── Example prompts ───────────────────────────────────────────────────
        gr.Examples(
            examples=[
                ["Roast me as my digital twin would 😂"],
                ["If I were Batman, how would I approach my DBMS exam?"],
                ["I have 8.4 CGPA after 21 subjects, I want 9.0 — is it even possible?"],
                ["What should I study today and in what order?"],
                ["Give me the hardest DSA problem for today."],
                ["Why should a startup hire a 2nd year like me?"],
                ["What is the weather in Bangalore right now?"],
                ["Calculate: sin(pi/4) * factorial(6) + log10(1000)"],
                ["Search Wikipedia: what is transformer architecture?"],
                ["Explain dynamic programming like I'm 10 years old."],
            ],
            inputs=msg_box,
            label="💡 Try these:"
        )

        # ── Wire up interactions ──────────────────────────────────────────────
        def respond(user_msg, history):
            if not user_msg.strip():
                return history, ""
            reply   = chat_fn(user_msg, history)
            history = history + [(user_msg, reply)]
            return history, ""

        def clear_chat():
            conversation_history.clear()
            return [(None, INTRO_MESSAGE)], ""

        # Mode buttons auto-fill the textbox
        btn_bg.click(lambda: "Tell me about yourself and your background.", outputs=msg_box)
        btn_calc.click(lambda: "Calculate: sqrt(256) + factorial(5)", outputs=msg_box)
        btn_study.click(lambda: "What should I study today?", outputs=msg_box)
        btn_weather.click(lambda: "What is the weather in Bangalore right now?", outputs=msg_box)
        btn_dsa.click(lambda: "Give me a DSA problem to practice today.", outputs=msg_box)
        btn_chat.click(lambda: "Tell me something interesting about AI.", outputs=msg_box)

        send_btn.click(respond,  [msg_box, chatbot], [chatbot, msg_box])
        msg_box.submit(respond,  [msg_box, chatbot], [chatbot, msg_box])

    demo.launch(debug=True)


launch_ui()
```
